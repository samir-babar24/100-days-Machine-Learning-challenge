```python
sam['Ground']=sam['Ground'].fillna(sam['Ground'].mode()[0])
```

```python
fea=sam.drop(columns=['Ground'])
tar=sam['Ground']
```

```python
xtr,xte,ytr,yte=train_test_Split=train_test_split(fea,tar,test_size=0.2)
```

```python
cat_cols = xtr.columns[:3].tolist()
num_cols = xtr.columns[3:6].tolist()
```

```python
oe=OneHotEncoder(handle_unknown='ignore')
sc=StandardScaler()
iti=IterativeImputer(max_iter=12,random_state=30)

pipe=Pipeline([('t2',sc),('t3',iti)])




ct=ColumnTransformer(
    transformers=[
       ('t1',oe,cat_cols),
       ('t4',pipe,num_cols)],remainder='passthrough')

```

```python
xtr_new=ct.fit_transform(xtr)
xte_new=ct.transform(xte)
```
