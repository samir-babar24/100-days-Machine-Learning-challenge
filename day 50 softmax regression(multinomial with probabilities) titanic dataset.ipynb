```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score
```

```python
df=sns.load_dataset('titanic')
```

```python
df.columns
```

```python
df.head(5)
```

```python
new_df=df.drop(columns=['who', 'adult_male', 'deck', 'embark_town',
       'alive', 'alone','survived', 'pclass', 'sex','sibsp', 'parch','embarked'])
```

```python
new_df
```

```python
new_df.isnull().sum()
```

```python
new_df.dropna(inplace=True)
```

```python
new_df.info()
```

```python

label=LabelEncoder()

new_df['class']=label.fit_transform(new_df['class'])



```

```python
new_df
```

```python
new_df.isnull().sum()
```

```python
x=new_df.drop(columns=['class'])
y=new_df['class']
```

```python
xtr,xte,ytr,yte=train_test_split(x,y,test_size=0.3,random_state=40)
```

```python

xtr
```

```python
ytr
```

```python

model=LogisticRegression(multi_class='multinomial')
```

```python
model.fit(xtr,ytr)
```

```python
pred=model.predict(xte)
print(pred)
```

```python
accuracy=accuracy_score(yte,pred)
print(accuracy)
```

```python
user=np.array([40,30.40])
results=model.predict(user.reshape(1,-1))
```

```python
print(results)
```

```python
each_result=model.predict_proba(user.reshape(1,-1))
print(each_result)
```
