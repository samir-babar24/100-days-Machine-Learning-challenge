```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
```

```python
sam=pd.read_csv("/content/CAR DETAILS FROM CAR DEKHO.csv")
sam
```

```python
sam.drop(columns=['name'])
```

```python
fea=sam.drop(columns=['selling_price'])
tar=sam['selling_price']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(fea,tar,test_size=0.2,random_state=30)
```

```python
transformer=ColumnTransformer(
    transformers=[
        ('t1',OneHotEncoder(sparse_output=False,drop='first'),['seller_type','transmission','fuel']),
        ('t2',OrdinalEncoder(categories=[['First Owner', 'Second Owner', 'Fourth & Above Owner',
       'Third Owner', 'Test Drive Car']]),['owner']),
        ('t3',StandardScaler(),['year','km_driven'])
    ],
    remainder='passthrough'
)
```

```python
x_train_transformmed=transformer.fit_transform(xtrain)
xtest_transformmed=transformer.transform(xtest)
```
