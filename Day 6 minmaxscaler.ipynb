```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
```

```python
sam=sns.load_dataset('tips')
inp=sam.iloc[:,0:2]

out=sam.iloc[:,-1]
out
inp
```

```python
x_train,x_test,y_train,y_test=train_test_split(inp,out,test_size=0.2,random_state=20)
scaler=MinMaxScaler()
scaler.fit(x_train)

x_train_scaled=scaler.transform(x_train)
x_test_scaled=scaler.transform(x_test)
pd_scaled_train=pd.DataFrame(x_train_scaled)
pd_scaled_test=pd.DataFrame(x_test_scaled)
pd_scaled_train.head()
```
