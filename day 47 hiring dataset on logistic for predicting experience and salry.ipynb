```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("hiring.csv")

df.head()
```

```python
word_to_num = {
    'zero': 0, 'one': 1, 'two': 2, 'three': 3, 'four': 4, 'five': 5,
    'six': 6, 'seven': 7, 'eight': 8, 'nine': 9, 'ten': 10
}
```

```python
df['experience'] = df['experience'].replace(word_to_num)
df['experience'] = pd.to_numeric(df['experience'], errors='coerce')

```

```python
imputer = SimpleImputer(strategy='mean')
X = imputer.fit_transform(df[['test_score(out of 10)', 'interview_score(out of 10)', 'salary($)']])


```

```python
le = LabelEncoder()
y = le.fit_transform(df['experience'])

```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X, y)

```

```python
accuracy_score=model.score(X,y)
print(accuracy_score)
```

```python
test = float(input())
interview = float(input())
salary = float(input())
```

```python
input_data = np.array([[test, interview, salary]])
predicted_label = model.predict(input_data)
predicted_experience = le.inverse_transform(predicted_label)
```
