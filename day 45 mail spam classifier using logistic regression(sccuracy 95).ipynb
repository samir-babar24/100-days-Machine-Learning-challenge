```python

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('mail_data.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.tail()
```

```python
df.shape
```

```python
encode=OneHotEncoder(sparse_output=False)
```

```python
df['Category']=encode.fit_transform(df[['Category']])
```

```python
df['Category']
```

```python
df.head()
```

```python
fea=df['Message']
tar=df['Category']
```

```python
xtr,xte,ytr,yte=train_test_split(fea,tar,test_size=0.3,random_state=20)
```

```python
#usinng the vectorizer here
vec=TfidfVectorizer()
xtr_new=vec.fit_transform(xtr)
xte_new=vec.transform(xte)
```

```python
print(xtr_new)
```

```python
model=LogisticRegression()
```

```python
model.fit(xtr_new,ytr)
```

```python
pred_tr=model.predict(xtr_new)
pre_te=model.predict(xte_new)
```

```python
#accuracy on the training data
acc=accuracy_score(pred_tr,ytr)
print("accuracy on training:" ,acc*100)
```

```python
#accuracy on testing data
acc1=accuracy_score(pre_te,yte)
print("accuracy on testing :", acc1*100)
```
