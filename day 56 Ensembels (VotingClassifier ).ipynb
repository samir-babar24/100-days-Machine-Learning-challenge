```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split,cross_val_score
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import VotingClassifier
```

```python
df=sns.load_dataset('exercise')
```

```python
df
```

```python
df.isnull().mean()
```

```python
df=df.drop(columns=['time'])
```

```python
df.info()
```

```python
le=LabelEncoder()
```

```python
df['kind'] = le.fit_transform(df['kind'])
df['diet'] = le.fit_transform(df['diet'])
```

```python
df
```

```python
x=df.drop(columns=['kind'])
y=df['kind']
```

```python
xtr,xte,ytr,yte=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
clf1=LogisticRegression()
clf2=RandomForestClassifier()
clf3=KNeighborsClassifier()
```

```python
estimators=[('l1',clf1),('l2',clf2),('l3',clf3)]
```

```python
for estimator in estimators:
    z = cross_val_score(estimator[1], x, y, cv=10, scoring='accuracy')
    print(estimator[0], np.round(np.mean(z), 2))
```

```python
vc=VotingClassifier(estimators=estimators)
z1=cross_val_score(vc,x,y,cv=10,scoring='accuracy')
print(np.mean(z1))
```

```python
vc_soft=VotingClassifier(estimators=estimators,voting='soft')
z1=cross_val_score(vc_soft,x,y,cv=10,scoring='accuracy')
print("\n\n\n", np.mean(z1))
```

```python
for i in range (1,4):
  for j in range (1,4):
    for k in range (1,4):
      vc1=VotingClassifier(estimators=estimators,weights=[i,j,k])
      cross=cross_val_score(vc1,xtr,ytr,scoring='accuracy',cv=4)
      print(f"for i={i}, j={j}, k={k} -> Accuracy:", np.round(np.mean(cross), 2))
```

```python
vc_final=VotingClassifier(estimators=estimators,voting='soft',weights=[1,1,1])
```

```python
model=vc_final.fit(xtr,ytr)
```

```python
predict=model.predict(xte)
```

```python
calculated_accuracy = accuracy_score(yte, predict)


print("accuracy=", calculated_accuracy)
```
