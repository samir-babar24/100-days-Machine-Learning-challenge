```python
#machinenlearning model without pipelines
import pandas as pd
import numpy as np
import seaborn as sns

from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
df=sns.load_dataset('titanic')
df.sample(5)
```

```python
#checking and removing unnecessary columns
df.isnull().sum()
sam=df.drop(columns=['deck','sex','alone','who','sibsp','parch','adult_male'])
```

```python
sam.sample(5)
```

```python
sam.isnull().sum()
```

```python
fea=sam.drop(columns='alive')
tar=sam['alive']
```

```python
#training ,testing and spliting of a data

xtr,xte,ytr,yte=train_test_split(fea,tar,random_state=30,test_size=0.3)
```

```python
#using a simple imputer to fill missing values
se_embarktown=SimpleImputer(strategy='most_frequent')
se_embarked=SimpleImputer(strategy='most_frequent')
se_age=SimpleImputer(strategy='mean')

xtr_embarktown=se_embarktown.fit_transform(xtr[['embark_town']])
xte_embarktown=se_embarktown.transform(xte[['embark_town']])

xtr_embarked=se_embarked.fit_transform(xtr[['embarked']])
xte_embarked=se_embarked.transform(xte[['embarked']])

xtr_age=se_age.fit_transform(xtr[['age']])
xte_age=se_age.transform(xte[['age']])

```

```python

#using of one hot encoder
ohe=OneHotEncoder(sparse_output=False,handle_unknown='ignore')
xtr_embarktown=ohe.fit_transform(xtr_embarktown)
xte_embarktown=ohe.transform(xte_embarktown)

xtr_embarked=ohe.fit_transform(xtr_embarked)
xte_embarked=ohe.transform(xte_embarked)
```

```python
xtr_transformmed=xtr.drop(columns=['embark_town','embarked','age'])
xte_transformmed=xte.drop(columns=['embark_town','embarked','age'])
```

```python
xtrain=np.concatenate((xtr_transformmed,xtr_embarked,xtr_embarktown,xtr_age),axis=1)
xtest=np.concatenate((xte_transformmed,xte_embarked,xte_embarktown,xte_age),axis=1)
```

```python
Model=DecisionTreeClassifier()
fitting=Model.fit(xtrain,ytr)
prediction=Model.predict(xtest)
```
