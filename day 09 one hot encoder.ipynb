```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder,StandardScaler,OrdinalEncoder
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
```

```python
fea=df.drop(columns=['selling_price'])
tar=df['selling_price']
xtrain,xtest,ytrain,ytest=train_test_split(fea,tar,test_size=0.2,random_state=30)
```

```python
#using nomina/onehot encoder
ohe=OneHotEncoder(sparse_output=False,drop='first')#here sparse_output  means to only store values without keeping zeros,and dropping first means have to avoid friat column to avoid dmulticolineariity
ohex_train=ohe.fit_transform(xtrain[['fuel','seller_type','transmission']])
ohex_test=ohe.transform(xtest[['fuel','seller_type','transmission']])

```

```python
pd.DataFrame(ohex_test)
```

```python
#now we are going for a ordinal data lets see
oe=OrdinalEncoder(categories=[[ 'Test Drive Car','Fourth & Above Owner','Third Owner', 'Second Owner','First Owner']])
oex_train=oe.fit_transform(xtrain[['owner']])
oex_test=oe.transform(xtest[['owner']])
```

```python
pd.DataFrame(oex_train)
```

```python
#using of standars scaler
st=StandardScaler()
stx_train=st.fit_transform(xtrain.drop(columns=['fuel','seller_type','transmission','owner','name']))
stx_test=st.transform(xtest.drop(columns=['fuel','seller_type','transmission','owner','name']))
```

```python
pd.DataFrame(stx_train)
```

```python
#combining all tranformation
x_train_transformmed=pd.concat([pd.DataFrame(ohex_train),pd.DataFrame(stx_train),pd.DataFrame(oex_train)],axis=1)
x_test_transformmes=pd.concat([pd.DataFrame(ohex_test),pd.DataFrame(stx_test),pd.DataFrame(oex_test)])
```
