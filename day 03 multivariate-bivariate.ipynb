```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
train=pd.read_csv('/content/train.csv')
train.head()
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
#scatter
scatter=sns.scatterplot(x="total_bill",y="tip",data=tips,hue="sex")
scatter
```

```python
#bar
bar=sns.barplot(x="species",y="petal_length",data=iris,color="red")
```

```python
#box
box=sns.boxplot(x="fare",y="sex",data=train)
```

```python
#dist
dist=sns.displot(x='total_bill',data=tips,hist=False,kde=True)
```
