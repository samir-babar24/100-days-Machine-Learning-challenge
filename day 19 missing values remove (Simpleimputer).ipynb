```python
sam=df.drop(columns=['Date', 'Location'	,'Operator','Time','Flight #','Registration', 'cn/In', 'Aboard',  'Ground'])
```

```python
fea=sam.drop(columns=['Fatalities'])
tar=sam['Fatalities']

xtrain,xtest,ytrain,ytest=train_test_split(fea,tar,test_size=0.2)
```

```python
ct=ColumnTransformer(
    transformers=[
        ('t1',SimpleImputer(strategy='most_frequent'),['Route','Type','Summary'])
    ],
    remainder='passthrough'
)
```

```python
xtraintrans=ct.fit_transform(xtrain)
xtesttrans=ct.transform(xtest)
```
