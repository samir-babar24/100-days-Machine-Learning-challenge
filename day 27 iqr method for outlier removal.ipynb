```python
#Data Preprocessing
#fiiling missing values
#age and deck
newdt=loadinf_dt
newdt['age'].fillna(newdt['age'].mean(),inplace=True)

# Check if 'unknown' is already a category
if 'unknown' not in newdt['deck'].cat.categories:
  # Add 'unknown' to the categories of the 'deck' column if it doesn't exist
  newdt['deck'] = newdt['deck'].cat.add_categories('unknown')

# Now you can fill the missing values with 'unknown'
newdt['deck'].fillna('unknown', inplace=True)
newdt['embark_town'].fillna('sanfransisco',inplace=True)
newdt['embarked'].fillna('S',inplace=True)

newdt
```

```python
newdt.isnull().sum()
```

```python
box1=sns.boxplot(x=newdt['fare'],color="maroon")
plt.xlabel('fare')
plt.title('boxplot with outliers')
plt.show()
```

```python
#here handling outliers
Q1=newdt['fare'].quantile(0.25)
Q3=newdt['fare'].quantile(0.75)
IQR=Q3-Q1

#setting th bounds

lowerbound= Q1-1.5*IQR
upperbound= Q3+1.5*IQR

#selecting region of 0.25 quintile to 0.75 quintile
datasetfiltered =newdt[(newdt['fare']>=lowerbound) & (newdt['fare']<=upperbound)]
meannn=datasetfiltered['fare'].mean()
#replacing all values which are out ot the ).25 quintile and 0.75 by meannn
newdt.loc[(newdt['fare']<lowerbound) | (newdt['fare']>upperbound),'fare']=meannn

#new ploting boxplot without outliers
box2=sns.boxplot(x=newdt['fare'],color="maroon")
plt.xlabel('fare')
plt.title('boxplot without outliers')
plt.show()
```
