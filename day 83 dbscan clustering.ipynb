```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import dbscan

```

```python
x, y_true = make_moons(n_samples=300, noise=0.1, random_state=42)
```

```python

print("first 5 samples of x:")
print(x[:5])
print("corresponding true labels:")
print(y_true[:5])
```

```python

plt.figure(figsize=(8, 6))
plt.scatter(x[:, 0], x[:, 1], c=y_true, cmap='viridis', s=50, edgecolor='k')
plt.title("raw data visualization - make_moons")
plt.xlabel("feature 1")
plt.ylabel("feature 2")
plt.show()
```

```python

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(x[:, 0], bins=20, color='skyblue', edgecolor='k')
plt.title("feature 1 distribution")
```

```python

plt.subplot(1, 2, 2)
plt.hist(x[:, 1], bins=20, color='salmon', edgecolor='k')
plt.title("feature 2 distribution")
plt.tight_layout()
plt.show()
```

```python
core_samples_mask, clusters = dbscan(X=x, eps=0.2, min_samples=5)
```

```python

plt.figure(figsize=(8, 6))
plt.scatter(x[:, 0], x[:, 1], c=clusters, cmap='coolwarm', s=50, edgecolor='k')
plt.title("dbscan clustering - basic example")
plt.xlabel("feature 1")
plt.ylabel("feature 2")
plt.show()
```

```python

print("unique clusters found by dbscan:", np.unique(clusters))
print("number of points in each cluster:")
for c in np.unique(clusters):
    print(f"cluster {c}: {(clusters == c).sum()} points")
```
